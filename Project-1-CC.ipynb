### Challenge 1: Largest Palindrome
A palindromic number reads the same both ways. For example, 1234321 is a palindrome. The largest palindrome made from the product of two two-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two three-digit numbers.

```python
palindromes_list = []
factor1_list = []
factor2_list = []
# Start by creating three empty lists, one for the palindromes and 2 for the factors making up each palindrome.
for factor1 in range(100,1000):
    for factor2 in range(100,factor1+1):
        # We need to iterate twice through all the three digit numbers to get every combination of two.
        # The second range only goes up through 'factor1' so we test each combo once instead of twice.
        # i.e. if we check 2 x 3, we don't need to also check 3 x 2.
        # This isn't strictly necessary, but I suppose it could cut down on processing time.
        product = factor1 * factor2
        reverse_product = int(str(product)[::-1])
        # Here we calculate the 'reverse' of the product, but we need to make the product into a string first.
        # We then convert the reversed string back into an integer.
        if product == reverse_product:
            palindromes_list.append(product)
            factor1_list.append(factor1)
            factor2_list.append(factor2)
            # If the product is a palindrome, we append the product and the two component factors to their respective lists.

answer = max(palindromes_list)
# The final answer is the biggest product in the list of palindromes.
# I had considered iterating backwards through the list, starting with 999 x 999 and working down to avoid checking all combinations.
# Unfortunately, it can be hard to tell with certainty that every subsequent product is less than the current one (i.e. 5x5 > 6x4).

counter = 0
index = 0
for i in palindromes_list:
    if i == answer:
        index = counter
    counter += 1
# This whole chunk is just to pick out the component factors of the 'winning' palindrome.
# It is not strictly necessary, but I wanted to check that the math actually worked, and it made a nicer output sentence.
    
print('The largest palindrome made from the product of two three-digit numbers is '+str(answer)+' = '+str(factor1_list[index])+' x '+str(factor2_list[index])+'.')  
# This line composes the output sentence using the 'winning' palindrome and the two component factors.
```


### Challenge 2: Summation of Primes
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. Find the sum of all the primes below 2,000.

```python
prime_list = [2]
prime_sum = 2
max_range = 2000
# I started the list of primes and the sum of primes with 2 'pre-loaded' because, using the methodology below, 2 is a tricky number to check.
# The methodology checks whether any number from 2 through one less than the number divides evenly into the number in question.
# Creating the list of primes is not strictly necessary, but it is helpful for checking.
# I created the max_range variable because this question is similar to the next one, and I wanted to copy the code as much as possible.
for number in range(3,max_range):
    counter = 0
    for i in range(2,number):
        if number % i == 0:
            counter += 1
            # Counter goes up each time a number is evenly divisible into the number in question.
    if counter == 0:
        prime_list.append(number)
        prime_sum += number
        # Only if the counter remained at zero will the number be added to the sum and list of primes.
        
print('The sum of all the primes below '+str(max_range)+' is '+str(prime_sum)+'.')
# This line composes the output sentence using the sum of primes and the max_range.
```

### Challenge 3: Multiples of 3 and 5
If we list all of the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6, and 9. The sum of these multiples is 23. Find the sum of all the multiples of 3 and 5 below 1,000.

```python
factor_list = []
factor_sum = 0
max_range = 1000
for number in range(1,max_range):
    if (number % 3 == 0) or (number % 5 == 0):
        factor_list.append(number)
        factor_sum += number

print('The sum of all the multiples of 3 and 5 below '+str(max_range)+' is '+str(factor_sum)+'.')
# This question is similar to the prior one.
# The only difference is that the methodology adds to the list and sum if the number divides evenly into 3 or 5.
# Again, the creation of the list is not strictly necessary but helps for checking.
```

# Bonus Challenge: FizzBuzz Extreme
Find or develop three different solutions to the following problem, in Python, that are as different from one another as possible. Which one do you prefer? Why? In what ways are the alternative solutions better?

Write a program that prints all of the numbers from 1 to 100. For multiples of 3, instead of the number, print "Fizz;" for multiples of 5, print "Buzz." For numbers that are multiples of both 3 and 5, print "FizzBuzz."

```python
max_range = 100

# Method 1
for i in range(1,max_range+1):
    if (i % 3 == 0):
        x = 'Fizz'
    else:
        x = ''
    if (i % 5 == 0):
        y = 'Buzz'
    else:
        y = ''
    print(i,x+y)
""" This was my solution from class. I like it because it only uses 2 'if' statements and captures the "heart" of the matter:
    If it's divisble by 3, add 'fizz'. If it's divisible by 5, add 'buzz'.
    If it's divisble by both, the situation is already taken care of. """

# Method 2
for i in range(1,max_range+1):
    if (i % 3 == 0) and (i % 5 == 0):
        print('FizzBuzz')
    elif (i % 3 == 0):
        print('Fizz')
    elif (i % 5 == 0):
        print('Buzz')
    else:
        print(i)
""" This is probably the first solution that would come to mind.
    It goes through the entire universe of scenarios and discretely provides the output for each.
    This would probably be the easiest code for someone to pick up and understand what it's doing. """

# Method 3
for i in range(1,max_range+1):
    if (i % 3 == 0) and (i % 5 == 0):
        x = 'Fizzbuzz'
    elif (i % 3 == 0):
        x = 'Fizz'
    elif (i % 5 == 0):
        x = 'Buzz'
    else:
        x = ''
    print(i,x)
""" One thing I don't like about Method 2 is that it doesn't print the number on a fizz/buzz number.
    This method combines 1 and 2. """
```
